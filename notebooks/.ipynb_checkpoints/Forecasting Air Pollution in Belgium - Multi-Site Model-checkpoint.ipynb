### Data Exploration

Let's use **pandas_profiling** to inspect the data. The pandas_profiling report learns us the following:

* There are 6 constant variables. We can remove these from the data set.
* No missing values exist, so probably we will not need to apply imputation.
* **AirPollutionLevel** has some zeroes, but this could be perfectly normal. On the other hand, this variables has some extreme values, which might be incorrect recordings of the air pollution.

* **DataCapture**: Proportion of valid measurement time relative to the total measured time (time coverage) in the averaging period, expressed as percentage. Almost all rows have about 100% of valid measurement time. Some rows have a DataCapture that is slightly lower than 100%.
* **DataCoverage**: Proportion of valid measurement included in the aggregation process within the averaging period, expressed as percentage. In this data set we have a minimum of 75%. According to the [definition of this variable](https://www.eea.europa.eu/data-and-maps/data/aqereporting-2/be) values below 75% should not be included for air quality assessments, which explains why these rows are not present in the data set. 
* **TimeCoverage**: highly correlated to DataCoverage and will be removed from the data.
* **UnitOfAirPollutionLevel**: 423 rows have a unit of *count*. To have a consistent target variable we will remove the records with this type of unit.
* **DateTimeBegin** and **DateTimeEnd**: the histogram does not provide enough detail here. This needs to be analyzed further.

#### Missing timesteps on daily aggregation level
As we can see below, **not all SamplingPoints have data for all DatetimeBegin in the three-year period**. These are most likely days where the DataCoverage variable was below 75%. So on these days we do not have sufficient valid measurements. Later in this notebook we will use measurements on prior days to predict the pollution on the current day. 

To have similarly sized timesteps, we will need to insert rows for the missing DatetimeBegin per SamplingPoint. We will **insert the measurement data of the next day with valid data**. 

Secondly, we will **remove the SamplingPoints with too many missing timesteps**. Here we will take an arbitrary number of 1.000 timesteps as the minimum number of required timesteps.

```python
ser_avail_days = agg_ts.groupby('SamplingPoint').nunique()['DatetimeBegin']
plt.figure(figsize=(8,4))
plt.hist(ser_avail_days.sort_values(ascending=False))
plt.ylabel('Nb SamplingPoints')
plt.xlabel('Nb of Unique DatetimeBegin')
plt.title('Distribution of Samplingpoints by the Nb of available measurement days')
plt.show()
```

### Data Preparation
#### Data Cleaning
Based on the data exploration, we will do the following to clean the data:
* Keeping only records with DataAggregationProcss of P1D
* Removing records with UnitOfAirPollutionLevel of count
* Removing unary variables and other redundant variables
* Removing SamplingPoints which have less than 1000 measurement days

```python
df = agg_ts.loc[agg_ts.DataAggregationProcess=='P1D', :] 
df = df.loc[df.UnitOfAirPollutionLevel!='count', :]
df = df.loc[df.SamplingPoint.isin(ser_avail_days[ser_avail_days.values >= 1000].index), :]
vars_to_drop = ['AirPollutant','AirPollutantCode','Countrycode','Namespace','TimeCoverage','Validity','Verification','AirQualityStation',
               'AirQualityStationEoICode','DataAggregationProcess','UnitOfAirPollutionLevel', 'DatetimeEnd', 'AirQualityNetwork',
               'DataCapture', 'DataCoverage']
df.drop(columns=vars_to_drop, axis='columns', inplace=True)
```

#### Inserting rows for the missing timesteps
For each SamplingPoint, we will first insert (empty) rows for which we do not have a DatetimeBegin. This can be done by creating a complete multi-index with all SamplingPoints and over the range between the minimum and maximum DatetimeBegin. Then, **reindex** will insert the missing rows but with NaN for the columns. 

Secondly, we use **bfill** and specify to impute the missing values with values of the next row with valid data. The bfill method is applied on a groupby object to limit the backfilling within the rows of each SamplingPoint. That way we do not use values of another SamplingPoint to fill in the missing values.

A samplepoint to test whether this operation worked correctly is *SPO-BETR223_00001_100* for the date *2013-01-29*.

```python
dates = list(pd.period_range(min(df.DatetimeBegin), max(df.DatetimeBegin), freq='D').values)
samplingpoints = list(df.SamplingPoint.unique())

new_idx = []
for sp in samplingpoints:
    for d in dates:
        new_idx.append((sp, np.datetime64(d)))

df.set_index(keys=['SamplingPoint', 'DatetimeBegin'], inplace=True)
df.sort_index(inplace=True)
df = df.reindex(new_idx)
#print(df.loc['SPO-BETR223_00001_100','2013-01-29'])  # should contain NaN for the columns

df['AirPollutionLevel'] = df.groupby(level=0).AirPollutionLevel.bfill().fillna(0)
#print(df.loc['SPO-BETR223_00001_100','2013-01-29'])  # NaN are replaced by values of 2013-01-30
print('{} missing values'.format(df.isnull().sum().sum()))
```

```python
plt.figure(figsize=(14,8))
viz_df = df.loc['SPO-BETR223_00001_100']
plt.plot(viz_df.index, viz_df)
plt.show()
```

Allright, now we have a data set that is cleaned and does not contain any missing values. Can we immediately use this data set for training then? Not exactly. There are still some steps required before we can pass it to Keras for modeling.

* currently the data set cannot be considered as a **supervised learning problem**. For that we need to convert it to have input feature(s) and a target feature.
* we have time series for **multiple SamplingPoints**. We need to take into account that a time series is of a specific SamplingPoint.
* Keras accepts only **three-dimensional input data**. Therefore we need to reshape it into a 3D array.
* In a last step we need to **split between train and test data**.

#### Conversion to a supervised learning problem -  Lagged AirPollutionLevels
To predict the AirPollutionLevel at timestep t we will create lagged features. This can be done with the **shift** method in Pandas. By applying the shift method on a GroupBy object, we make sure not to shift AirPollutionLevels from one SamplingPoint to another. 

In this notebook we limit the number of shifts to 2. This results in missing lagged values for the first and second row of each SamplingPoint because we have no history before these timesteps. These rows are removed by applying the **dropna** method.

```python
def create_lagged_df(df, var_name, lag=1):
    if not type(df) == pd.DataFrame:
        raise TypeError('df must be a Pandas DataFrame')
    
    def rename_var(v, j):
        new_v = v + f'(t-{j})'
        return new_v

    for i in range(1, lag + 1):
        new_var_name = rename_var(var_name, i)
        df[new_var_name] = df.groupby(level=0)[var_name].shift(i)
            
    return df.dropna()

n_lag = 2
lagged_df = create_lagged_df(df, 'AirPollutionLevel', lag=n_lag)
```

#### Taking into account multiple SamplingPoints
As the training data is rather limited, we will not build a model for each SamplingPoint separately. Instead we will create dummy variables to represent for which SamplingPoint the row in the training data is.

```python
lagged_df.reset_index(level='SamplingPoint', inplace=True)
lagged_df = pd.get_dummies(lagged_df)
```

#### Split train, test and validation set
We split off a test set in order to evaluate the performance of the model. The test set will not be used during the training phase. We will also use a validation set that we can use during training. 

* train set : data of 2013
* validation set : data of 2014
* test set : data of 2015

By splitting the sets by year we can evaluate the model performance for all SamplingPoints. It allows us to use the dummy variables representing the SamplingPoints in the model.

```python
train = lagged_df.query('DatetimeBegin < "2014-01-01"')
valid = lagged_df.query('DatetimeBegin >= "2014-01-01" and DatetimeBegin < "2015-01-01"')
test = lagged_df.query('DatetimeBegin >= "2015-01-01"')

train.to_csv('../data/processed/train.csv')
valid.to_csv('../data/processed/valid.csv')
test.to_csv('../data/processed/test.csv')
```

#### Reshaping to 3D input data
In order to use the data as input for a LSTM layer in Keras we need to reshape the data. In Keras the input data needs to be a 3D array. These 3 dimensions represent the following:
* **samples** : each sequence represents a sample. So this will come down to the number of rows we have in our data.
* **timesteps** : each observation (or DatetimeBegin) in the sequence is a timestep. In our case this will be the number of lagged AirPollutionLevels in each row.
* **features** : the number of features per timestep. We look at the AirPollutionLevel of previous timesteps to predict the AirPollutionLevel at the current day. Additionally we have dummy variables for the SamplingPoints.

Before reshaping we convert the DataFrame to an array and split off the target variable and the input features.

```python
train = pd.read_csv('../data/processed/train.csv', header=0, index_col=0).values.astype('float32')
valid = pd.read_csv('../data/processed/valid.csv', header=0, index_col=0).values.astype('float32')
test = pd.read_csv('../data/processed/test.csv', header=0, index_col=0).values.astype('float32')

scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit(train)
train = scaler.transform(train)
valid = scaler.transform(valid)
test = scaler.transform(test)


train_X, train_y = train[:, 1:], train[:, 0]
valid_X, valid_y = valid[:, 1:], valid[:, 0]
test_X, test_y = test[:, 1:], test[:, 0]








train_X = train_X.reshape((train_X.shape[0], 1, train_X.shape[1]))
valid_X = valid_X.reshape((valid_X.shape[0], 1, valid_X.shape[1]))
test_X = test_X.reshape((test_X.shape[0], 1, test_X.shape[1]))
print('Train shape:',train_X.shape, train_y.shape)
print('Valid shape:',valid_X.shape, valid_y.shape)
print('Test shape:',test_X.shape, test_y.shape)
```

### Modeling
#### Recurrent Neural Networks
**Traditional neural networks have no memory**. Consequently they do not take into account previous input when processing the current input. In sequential data sets, like time series, the information of previous time steps is typically relevant for predicting something in the current step. So a **state** about the previous time steps needs to be maintained. 

In our case, the air pollution at time t might be influenced by the air pollution in previous timesteps.  So we need to take that into account. Recurrent Neural Networks or RNNs have an internal loop by which they maintain a state of previous timesteps. This state is then used for the prediction in the current timestep. The state is reset when a new sequence is being processed.

For an illustrated guide about RNNs you should definitely read the [article by Michael Nguyen](https://towardsdatascience.com/illustrated-guide-to-recurrent-neural-networks-79e5eb8049c9). 

```python
def plot_loss(history, title):
    plt.figure(figsize=(10,6))
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Validation')
    plt.title(title)
    plt.xlabel('Nb Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
```

```python
simple_rnn = Sequential()
simple_rnn.add(SimpleRNN(train_X.shape[2]))
simple_rnn.add(Dense(1))
simple_rnn.compile(loss='mae', optimizer=RMSprop())
simple_rnn_history = simple_rnn.fit(train_X, train_y, epochs=25, batch_size=32, validation_data=(valid_X, valid_y), verbose=0, shuffle=False)
plot_loss(simple_rnn_history, 'SimpleRNN - Train & Validation Loss')
```

#### Simple neural net

```python
train_X.reshape(-1,train_X.shape[2]).shape
```

```python
train_X_2D = train_X.reshape(-1,train_X.shape[2])
train_y_1D = train_y.reshape(-1,)

valid_X_2D = valid_X.reshape(-1,valid_X.shape[2])
valid_y_1D = valid_y.reshape(-1,)

simple_nn = Sequential()
simple_nn.add(Dense(45,input_dim=train_X.shape[2]))
simple_nn.add(Dense(24))
simple_nn.add(Dense(12))
simple_nn.add(Dense(1))
simple_nn.compile(loss='mae', optimizer=RMSprop())
simple_nn_history = simple_nn.fit(train_X_2D, train_y_1D, epochs=25, batch_size=32, validation_data=(valid_X_2D, valid_y_1D), verbose=0, shuffle=False)
plot_loss(simple_nn_history, 'Simple NN - Train & Validation Loss')
```

#### Long Short Term Memory Networks
**A RNN has a short memory**. It has difficulty to remember information of many timesteps ago. This occurs when the sequences are very long. In fact it is due to the **vanishing gradient problem**. The gradient are values that update the weights of a neural network. When you have many timesteps in your RNN the gradient for the first layers becomes very tiny. As a result the update of the weights of the first layers is negligable. This means that the RNN is not capable of learning what was in the early layers. 

So we need a way to carry the information of the first layers to later layers. LSTMs are better suited to take into account long-term dependencies. Michael Nguyen wrote an [excellent article](https://towardsdatascience.com/illustrated-guide-to-lstms-and-gru-s-a-step-by-step-explanation-44e9eb85bf21) that visually describes the workings of LSTMs and GRUs. 

#### Simple LSTM model
This is a simple model with the following architecture:
1. Input layer
2. Fully connected LSTM layer
3. Fully connected output layer

```python
simple_lstm = Sequential()
simple_lstm.add(LSTM(train_X.shape[2]))
simple_lstm.add(Dense(1))
simple_lstm.compile(loss='mae', optimizer=RMSprop())
simple_lstm_history = simple_lstm.fit(train_X, train_y, epochs=25, batch_size=32, validation_data=(valid_X, valid_y), verbose=0, shuffle=False)
plot_loss(simple_lstm_history, 'Simple LSTM - Train & Validation Loss')
```

#### Stacked LSTM model
In this model we will be stacking multiple (here two) LSTM layers. That way the model will learn other abstractions of input data over time. In other words, **representing the input data at different time scales**. 

To do that in Keras, we need to specify the parameter *return_sequences* in the LSTM layer preceding another LSTM layer.

```python
stacked_lstm = Sequential()
stacked_lstm.add(LSTM(train_X.shape[2], return_sequences=True))
stacked_lstm.add(LSTM(train_X.shape[2]))
stacked_lstm.add(Dense(1))
stacked_lstm.compile(loss='mae', optimizer=RMSprop())
stacked_lstm_history = stacked_lstm.fit(train_X, train_y, epochs=25, batch_size=32, validation_data=(test_X, test_y), verbose=0, shuffle=False)
plot_loss(stacked_lstm_history, 'Stacked LSTM - Train & Validation Loss')
```

### Evaluating Performance

```python
# make a prediction with the RNN
yhat_simple_rnn = simple_rnn.predict(test_X)

# make a prediction with the LSTM
yhat_simple_lstm = simple_lstm.predict(test_X)

# make a prediction with the LSTM
yhat_stacked_lstm = stacked_lstm.predict(test_X)

test_X = test_X.reshape((test_X.shape[0], 1*47))
```

```python
# make a prediction based on a naive baseline, being the air pollution of the previous timestep
yhat_naive = test_X[:,0].reshape(-1,1)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
def evaluate_rmse(yhat, test_y):
    # invert scaling for forecast
    inv_yhat = np.concatenate((yhat, test_X), axis=1)
    inv_yhat = scaler.inverse_transform(inv_yhat)
    inv_yhat = inv_yhat[:,0]
    # invert scaling for actual
    test_y = test_y.reshape((len(test_y), 1))
    inv_y = np.concatenate((test_y, test_X), axis=1)
    inv_y = scaler.inverse_transform(inv_y)
    inv_y = inv_y[:,0]
    # calculate RMSE
    rmse = np.sqrt(mean_squared_error(inv_y, inv_yhat))
    print('Test RMSE: %.3f' % rmse)
```

```python
evaluate_rmse(yhat_simple_rnn, test_y)
```

```python
evaluate_rmse(yhat_simple_lstm, test_y)
```

```python
evaluate_rmse(yhat_naive, test_y)
```

```python
evaluate_rmse(yhat_stacked_lstm, test_y)
```
